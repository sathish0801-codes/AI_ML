**EDA** stands for **Exploratory Data Analysis**.

In data science and machine learning, EDA is the critical initial phase where you analyze, summarize, and visualize a dataset to understand its underlying structure, patterns, and anomalies before applying any formal modeling or predictive algorithms.

Think of EDA as **"getting to know your data."** Instead of blindly feeding a dataset into a machine learning model, a data scientist uses EDA to look at the data from different angles to ensure it is clean, viable, and understood.

---

### Core Objectives of EDA

1. **Understand Data Structure & Integrity**
* Identify the size of the dataset (rows and columns).
* Check data types (e.g., ensuring numerical values aren't accidentally saved as text/strings).
* Spot missing values, duplicate records, or structural errors.


2. **Analyze Distributions**
* Determine how data spreads across variables using metrics like mean, median, standard deviation, and range.
* Look at the "shape" of data distributions (e.g., Is it a symmetrical bell curve, or is it heavily skewed to one side?).


3. **Discover Relationships & Correlations**
* Investigate how different features interact with one another. (e.g., Does a car's weight increase as its engine size increases?).
* Identify which features share a strong correlation with the target variable you want to predict.


4. **Detect Outliers**
* Find extreme values or anomalies that deviate significantly from the rest of the data. Outliers can heavily distort statistical models, so identifying them early is vital.


5. **Formulate Hypotheses & Inform Feature Engineering**
* EDA helps you decide how to handle missing data (imputation), whether you need to normalize/scale features, and which variables are worth keeping for the machine learning model.



---

### Common Tools and Techniques Used in EDA

Data scientists primarily rely on summary statistics and data visualizations during EDA:

* **Statistical Metrics:** Running operations like `.info()` or `.describe()` in Pandas to view data frames, count distributions, quartiles, and averages.
* **Univariate Visualizations (One Variable):** * *Histograms* and *Density Plots (KDE)* to see the distribution of continuous numbers.
* ***Count Plots*** or ***Bar Charts*** to see the frequency distribution of categorical items (like counting how many cars are from the USA vs. Japan).


* **Bivariate/Multivariate Visualizations (Multiple Variables):**
* ***Scatter Plots*** to map continuous variables against each other.
* ***Violin Plots*** or ***Box Plots*** to see how numerical features change across different categorical target groups.
* ***Correlation Heatmaps*** to instantly visualize linear relationships across the entire dataset matrix.
