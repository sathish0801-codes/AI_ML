````python
car_data.columns.tolist() #display only column names
````

````python
car_data.nunique().idxmax() #column has maximum number of unique values
````

````python
highest_unique_count = 0
column_name = ""

for col in car_data.columns:

    unique_count = car_data[col].nunique()

    if unique_count > highest_unique_count:
        highest_unique_count = unique_count
        column_name = col

print("Column:", column_name)
print("Unique Values:", highest_unique_count)
````
